```python
#Imports
import numpy as np
import matplotlib.pyplot as plt
```

![](img/1.png)

```python
A = np.array([3, -2])
B = np.array([-1, -4])
```

## a) A + B

```python
result_a = A+B
plt.title(f'|A+B|={result_a}')
plt.quiver(0, 0, A[0], A[1], scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(A[0], A[1], B[0], B[1], scale_units='xy', angles='xy', scale=1, color='g')
plt.quiver(result_a[0], result_a[1], scale_units='xy', angles='xy', scale=1, color='r')
plt.xlim(0, 4)
plt.ylim(-8, 0)
plt.grid()
plt.show()
```

## b) A - B

```python
#b) A-B
result_b = A-B
plt.title(f'|A-B|={result_b}')
plt.quiver(0, 0, A[0], A[1], scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(A[0], A[1], -B[0], -B[1], scale_units='xy', angles='xy', scale=1, color='g')
plt.quiver(result_b[0], result_b[1], scale_units='xy', angles='xy', scale=1, color='r')
plt.xlim(0, 5)
plt.ylim(-8, 5)
plt.grid()
plt.show()
```

## c) |A + B|

```python
result_c = np.linalg.norm(A+B)
print(f'|A+B| = {result_c}')
```

## d) |A - B|

```python
result_c = np.linalg.norm(A-B)
print(f'|A-B| = {result_c}')
```

## e)direccion A + B y A - B

```python
theta_a = -np.arccos((A+B)[0]/np.linalg.norm(A+B))
degrees = np.degrees(theta_a)
print(f'Direccion vector A + B = {degrees:.2f}°')
```

```python
theta_b = -np.arccos((A-B)[0]/np.linalg.norm(A-B))
degrees = np.degrees(theta_b)
print(f'Direccion vector A - B = {degrees:.2f}°')
```

![](img/2.png)

```python
A = np.array([-3, 4])
B = np.array([2, 3])
```

## a) A X B

```python
dot_vec = np.cross(A, B)
plt.title(f'A X B = {dot_vec}')
plt.quiver(0, 0, A[0], A[1], scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0, 0, B[0], B[1], scale_units='xy', angles='xy', scale=1, color='g')
plt.xlim(-4, 3)
plt.ylim(0, 5)
plt.grid()
plt.show()
```

## b) angulo entre A y B 

```python
mod_a = np.linalg.norm(A)
mod_b = np.linalg.norm(B)
dot_vec = np.dot(A, B)

theta = np.arccos(dot_vec/(mod_a*mod_b))
degrees = np.degrees(theta)
print(f'Angulo ente A X B = {degrees:.2f}°')
```

![](img/3.png)

```python
F = np.array([6, -2])
s = np.array([3, 1])
```

## A) W = F.s

```python
w = np.dot(F, s)
plt.title(f'W = F.s = {w} N.m')
plt.quiver(0, 0, F[0], F[1], scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0, 0, s[0], s[1], scale_units='xy', angles='xy', scale=1, color='g')
plt.xlim(0, 6)
plt.ylim(-3, 2)
plt.grid()
plt.show()
```

## b) angulo F y s

```python
mod_f = np.linalg.norm(F)
mod_s = np.linalg.norm(s)
dot_vec = np.dot(F, s)

theta = np.arccos(dot_vec/(mod_f*mod_s))
degrees = np.degrees(theta)
print(f'Angulo ente F.s = {degrees:.2f}°')
```
