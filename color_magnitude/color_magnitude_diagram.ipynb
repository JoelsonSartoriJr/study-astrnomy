```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits

#set size plot
plt.rcParams["figure.figsize"] = (20, 10)
```

```python
# Gerating Isochrones http://stev.oapd.inaf.it/cgi-bin/cmd
loga, logl, logte = np.loadtxt('data/isoc_z008.dat', usecols=(0,3,4), unpack=True)
```

```python
w7 = np.where(loga == 7.0)
w7_5 = np.where(loga == 7.5)
w8 = np.where(loga == 8.0)
w8_5 = np.where(loga == 8.5)
w9 = np.where(loga == 9.0)
```

```python
plt.plot(logte[w7], logl[w7], label='10 Myr')
plt.plot(logte[w8], logl[w8], label='100 Myr')
plt.plot(logte[w7_5], logl[w7_5], label='31.6 Myr')
plt.plot(logte[w8_5], logl[w8_5], label='316 2Myr')
plt.plot(logte[w9], logl[w9], label='1 Byr')

plt.xlabel('Log(Teff)')
plt.ylabel('Log L')
plt.axis([5.0, 3.5, 0, 6])
plt.title('Isochrones for star cluster with metalicity z = 0.008')
plt.legend()
plt.savefig('graph/cmdconstmetalicity.pdf')
```
