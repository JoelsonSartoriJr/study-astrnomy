```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits

#set size plot
plt.rcParams["figure.figsize"] = (15, 10)
```

```python
# Gerating Isochrones http://stev.oapd.inaf.it/cgi-bin/cmd
loga, logl, logte = np.loadtxt('data/isoc_z008.dat', usecols=(0,3,4), unpack=True)
```

```python
w7 = np.where(loga == 7.0)
w7_5 = np.where(loga == 7.5)
w8 = np.where(loga == 8.0)
w8_5 = np.where(loga == 8.5)
w9 = np.where(loga == 9.0)
```

```python
plt.plot(logte[w7], logl[w7], label='10 Myr')
plt.plot(logte[w8], logl[w8], label='100 Myr')
plt.plot(logte[w7_5], logl[w7_5], label='31.6 Myr')
plt.plot(logte[w8_5], logl[w8_5], label='316 2Myr')
plt.plot(logte[w9], logl[w9], label='1 Byr')

plt.xlabel('Log(Teff)')
plt.ylabel('Log L')
plt.axis([5.0, 3, 0, 6])
plt.title('Isochrones for star cluster with metalicity z = 0.008')
plt.legend()
plt.savefig('graph/cmdconstmetalicity.pdf')
```

```python
#Observational data using u.v
Mb1, Mv1 = np.loadtxt('data/isoc_z008.dat', usecols=(8,9), unpack=True)
Mb2, Mv2 = np.loadtxt('data/isoc_z0001.dat', usecols=(8,9), unpack=True)
Mb3, Mv3 = np.loadtxt('data/isoc_z0004.dat', usecols=(8,9), unpack=True)
Mb4, Mv4 = np.loadtxt('data/isoc_z004.dat', usecols=(8,9), unpack=True)
Mb5, Mv5 = np.loadtxt('data/isoc_z019.dat', usecols=(8,9), unpack=True)
Mb6, Mv6 = np.loadtxt('data/isoc_z030.dat', usecols=(8,9), unpack=True)
```

```python
loga1 = np.loadtxt('data/isoc_z008.dat', usecols=(0), unpack=True)
loga2 = np.loadtxt('data/isoc_z0001.dat', usecols=(0), unpack=True)
loga3 = np.loadtxt('data/isoc_z0004.dat', usecols=(0), unpack=True)
loga4 = np.loadtxt('data/isoc_z004.dat', usecols=(0), unpack=True)
loga5 = np.loadtxt('data/isoc_z019.dat', usecols=(0), unpack=True)
loga6 = np.loadtxt('data/isoc_z030.dat', usecols=(0), unpack=True)
```

```python
w71 = np.where(loga1 == 8.0)
w72 = np.where(loga2 == 8.0)
w73 = np.where(loga3 == 8.0)
w74 = np.where(loga4 == 8.0)
w75 = np.where(loga5 == 8.0)
w76 = np.where(loga6 == 8.0)
```

```python
plt.scatter(Mb1[w71] - Mv1[w71], Mv1[w71], label='z=0.008', s=1.8)
plt.scatter(Mb2[w72] - Mv2[w72], Mv2[w72], label='z=0.0001', s=1.8)
plt.scatter(Mb3[w73] - Mv3[w73], Mv3[w73], label='z=0.004', s=1.8)
plt.scatter(Mb4[w74] - Mv4[w74], Mv4[w74], label='z=0.04', s=1.8)
plt.scatter(Mb5[w75] - Mv5[w75], Mv5[w75], label='z=0.019', s=1.8)
plt.scatter(Mb6[w76] - Mv6[w76], Mv6[w76], label='z=0.030', s=1.8)
plt.title("Isochrones for star cluster with age 100 Myr")
plt.xlabel('B-V')
plt.ylabel('V')
plt.legend()
plt.savefig('graph/cmdconstage.pdf')
```
