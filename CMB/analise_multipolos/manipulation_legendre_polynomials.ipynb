```python

import numpy as np
from scipy.special import eval_legendre
from scipy.integrate import simps
import matplotlib.pyplot as plt
```

```python
X = np.linspace(-1, 1, 1000)
```

```python
for n in range(0, 5):
    y = eval_legendre(n, X)
    plt.plot(X, y, label=r'$P_{}(x)$')
    
plt.title('Legendre Polynomials')
plt.xlabel('x')
plt.ylabel(r'$P_n(x)$')
plt.grid(True)
plt.legend(loc='lower right')
plt.show()
```

```python

#orthogonality test
X = np.linspace(-1, 1, 100) 

P2 = eval_legendre(2, X)
P3 = eval_legendre(3, X)

print(f'Integral of P2 P3: {simps(P2*P3, X)}')
print(f'Integral of P2 P2: {simps(P2*P2, X)}')
print(f'Integral of P3 P3: {simps(P3*P3, X)}')
```

```python

# expansion legendre polynomials

X = np.linspace(-1, 1, 2000)

F = np.sin(10*X)*np.exp(-X)

plt.plot(X, F, 'ko', label='Original function')

f_new = np.zeros(len(X))
A = []
for n in range(0, 101):
    Pn = eval_legendre(n, X)
    an = (n+0.5) * simps(F*Pn, X)
    A.append(an)
    
    f_new += an*Pn
    if n%20 == 0:
        plt.plot(X, f_new, '--', label=f'F in n={n}')
        
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper left')
plt.grid()

plt.figure()
plt.xlabel('x')
plt.ylabel('$a_n$')
plt.title('powers')
plt.plot(A, 'bo-')
plt.grid()
plt.show()
```
