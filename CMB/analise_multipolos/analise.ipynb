```python
import math
import sympy as sy
import numpy as np
from sympy.functions import sin, cos
import matplotlib.pyplot as plt

from typing import func

plt.style.use('ggplot')
```

```python
x = sy.Symbol('x')
f = sin(x)
```

```python
def taylor(function, x0:float, n:int)->float:
    p =0
    for i in range(n+1):
        p += (function.diff(x, i).subs(x, x0))/(math.factorial(i))*(x-x0)**i
    return p
```

```python
xx = np.linspace(0, math.pi, 100)

for j in range(1,10,2):
    func = taylor(f, 0, j)
    lb = f'n = {j} : {func}'
    
    yy = []
    for value in xx:
        yy.append(func.subs(x, value))
    plt.plot(xx, yy, '--', label=lb)
    
plt.plot(xx, np.sin(xx), label='sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.title('Aproximacao em serie de Taylor')
plt.show()
```
