# Variacao do produto das funcoes

```python
x = np.linspace(-1, 1, 11000)
h0 = 0.0
dh = 0.05

y1 = f(x)
y1h = f(x+h0)
y1y1h = prod(x,h0)
```

```python
%matplotlib widget

fig, ax = plt.subplots()
plt.subplots_adjust(bottom=0.25)

l1, = plt.plot(x, y1, label='$f(x)$')
l2, = plt.plot(x, y1h, label='$f(x+h)$')
l3, = plt.plot(x, y1y1h, label='$f(x)*f(x+h)$')

plt.legend()
plt.grid()

axcolor = 'lightgoldenrodyellow'
axH = plt.axes([0.25, 0.1, 0.65, 0.08], facecolor=axcolor)
sH = Slider(axH, 'h', -1, 1, valinit=h0, valstep=dh)

def update(val):
    h = sH.val
    
    y1 = f(x)
    y1h = f(x+h)
    y1y1h = prod(x,h)
    
    l1.set_ydata(y1)
    l2.set_ydata(y1h)
    l3.set_ydata(y1y1h)
    
    fig.canvas.draw_idle()
    
sH.on_changed(update)
plt.show()
```

# Calculo da integral do produto

```python
%matplotlib widget

fig, ax = plt.subplots()
plt.subplots_adjust(bottom=0.25)

l1, = plt.plot(x, y1, label='$f(x)$')
l2, = plt.plot(x, y1h, label='$f(x+h)$')
l3, = plt.plot(x, y1y1h, label='$f(x)*f(x+h)$')

corr = integrate.quad(prod, -np.inf, np.inf, args=(h))[0]

tc = '$\int_{-\infty}^\infty f(x)f(x%+.03f) dx = %0.3f$'%(h, corr)
t1  = plt.text(.5, .98, tc, size=13, ha='center', va='top', transform=fig.transFigure,
                bbox=dict(boxstyle='round', ec=(1., 0.5, 0.5), fc=(1., 0.8, 0.8)))

plt.legend()
plt.grid()

axcolor = 'lightgoldenrodyellow'
axH = plt.axes([0.25, 0.1, 0.65, 0.08], facecolor=axcolor)
sH = Slider(axH, 'h', -1, 1, valinit=h0, valstep=dh)

def update(val):
    h = sH.val
    
    y1 = f(x)
    y1h = f(x+h)
    y1y1h = prod(x,h)
    
    l1.set_ydata(y1)
    l2.set_ydata(y1h)
    l3.set_ydata(y1y1h)
    
    corr = integrate.quad(prod, -np.inf, np.inf, args=(h))[0]
    tc = '$\int_{-\infty}^\infty f(x)f(x%+.03f) dx = %0.3f$'%(h, corr)
    t1.set_text(tc)
    
    fig.canvas.draw_idle()
    
sH.on_changed(update)
plt.show()
```

# Graph value integrante
