```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Tcmb = 2.728
dTD = 3.353e-3
dTA = 18e-6
n = 1000
```

```python
#Media
vT = Tcmb*np.ones(n)
```

```python
#Dipole
vdTD = np.arange(n)
vdTD = np.where(vdTD<n/2, Tcmb - dTD/2.0, Tcmb+dTD/2.0)
```

```python
#Anisotropias
vdTA = np.random.normal(0, dTA, n)
vdTA += vdTD
```

```python
plt.plot(vT, 'b-', label='Media')
plt.plot(vdTD, 'r-', label='Dipolo')
plt.plot(vdTA, 'k-', label='Anisotropias')

plt.ylim(0, 5)
plt.legend(loc='upper left')
plt.grid()
plt.show()
```
